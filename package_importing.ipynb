```python
#This is how to create new variables. Upon reaching class start creating the following variables

#1. Age category: Above 35, below 35
#2. Sex category: Male, female
#3. Occupation: Employed, not employed
sti['marital_status'] = None
sti.loc[(sti['A5MaritalStatus']=='1 single') | (sti['A5MaritalStatus']=='4 divorcee') | (sti['A5MaritalStatus']=='5 widowed'), 'marital_status']=0
sti.loc[(sti['A5MaritalStatus']=='2 married') | (sti['A5MaritalStatus']=='3 co-habiting'), 'marital_status']=1
sti['marital_status'] = sti['marital_status'].map({1: 'Live in Together', 0: 'single'})
print(sti['marital_status'])
    
sti['religion'] = None
sti.loc[(sti['A3Church'] =='2 apostolic') |(sti['A3Church'] =='3 methodist') |(sti['A3Church'] =='4 anglican')|(sti['A3Church'] =='5 pentecostal')|(sti['A3Church'] =='7 roman catholic'),'religion']=1
sti.loc[(sti['A3Church'] =='6 atheist') | (sti['A3Church'] =='8 other'), 'religion']=0
sti['religion'] = sti['religion'].map({1: 'Believers', 0: 'Non believers'})
print(sti['religion'])

sti['use_condom'] = None
sti.loc[sti['N12UseCondom']=='1 yes','use_condom'] = 1
sti.loc[sti['N12UseCondom']=='2 no','use_condom'] = 0
sti['use_condom'] = sti['use_condom'].map({1: 'yes', 0: 'no'})
print(sti['use_condom'])

sti['sti_infection'] = None
sti.loc[(sti['CaseStatus'] == 1) | (sti['CaseStatus'] == 3), 'sti_infection'] = 'Positive'
sti.loc[sti['CaseStatus'] == 2, 'sti_infection'] ='Negative' 
# Creating value labels for sti_infection
sti['sti_infection'] = sti['sti_infection'].map({'Positive': 1, 'Negative':0})
print(sti['sti_infection'])
##sti.sti_infection=sti.sti_infection.astype(int)~ changing from string to int


sti['agecat']=None
sti.loc[sti['A1Age']>35,'agecat']= 1
sti.loc[sti['A1Age']<=35,'agecat']=0 
sti['agecat'] = sti['agecat'].map({1: 'Above 35', 0:'Below 35'})
print(sti['agecat'])

sti['sex_partner'] = None
sti.loc[sti['SexPartner1year']==0, 'sex_partner']=0
sti.loc[sti['SexPartner1year']==1, 'sex_partner']=1
sti['sex_partner'] = sti['sex_partner'].map({1: 'Have partner', 0: 'Lack partner'})
print(sti['sex_partner'])

sti['sexcat'] = None
sti.loc[sti['Sex']=='Male', 'sexcat'] = 0
sti.loc[sti['Sex']=='Female', 'sexcat'] = 1
sti['sexcat'] = sti['sexcat'].map({1: 'Female', 0: 'Male'}) 
print(sti['sexcat'])

sti['occupation'] = None
sti.loc[(sti['A2Occupation'] == '2 informal') | (sti['A2Occupation'] == '3 formal'), 'occupation'] = 1
sti.loc[(sti['A2Occupation'] == '1 unemployed') | (sti['A2Occupation'] == '4 student'), 'occupation'] = 0
sti['occupation'] = sti['occupation'].map({1: 'employed', 0: 'not employed'})
print(sti['occupation'])

sti['education'] = None
sti.loc[(sti['A4LevelOfEducation']=='2 primary') | (sti['A4LevelOfEducation']=='1 none'),'education']=0
sti.loc[(sti['A4LevelOfEducation']=='3 secondary') | (sti['A4LevelOfEducation']=='4 tertiary'),'education']=1
sti['education'] = sti['education'].map({1: 'secondary and above', 0: 'primary and below'})
print(sti['education'])
sti.education.unique()

```

```python
from  scipy.stats import chi2_contingency

#create a contingency table
contingency_table = pd.crosstab(sti['sex_partner'], sti['sti_infection'])
print(contingency_table)

# perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

#print results
print("chi-square Test for: sex_partner vs sti_infection")
print(f"chi2 statistic: {chi2:.3f}")
print(f"p-value: {p:.4f}")
print(f"Degree of freedom: {dof}")
print("Expected Frequencies Table:")
print(expected)

#Interpretation
if p < 0.05:
    print("significant association (p< 0.05)")
else:
    print("No significant association (p >= 0.05)")
```

```python
# List of categorical variables

cat_vars = [
    "agecat", "sexcat","education","occupation","religion","marital_status","use_condom","sex_partner"]

#Loop over each variable and perform chi-square test
for var in cat_vars:
    print(f"\nChi-square Test for: {var}")
    contingency_table = pd.crosstab(sti[var], sti["sti_infection"])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    print(f"Chi2 statistic: {chi2:.3f}, p-value: {p:.4f}, Degrees of freedom: {dof}")
    if p < 0.05:
        print("Result: significant association (p< 0.05)") 
    else:
        print("Result: No significant association (p >= 0.05)")
```
